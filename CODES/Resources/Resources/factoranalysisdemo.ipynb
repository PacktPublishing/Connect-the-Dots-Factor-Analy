```python

import pandas as pd 
import numpy as np
```

```python
combinedFile = '/Users/swethakolalapudi/Desktop/Reg_Data/PricesHistorical3.csv'

data=pd.read_csv(combinedFile ,sep=",")
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
data.head()
```

```python
data = data.sort_values(['Date'], ascending=[True])
data.head()
```

```python
returns = data[[key for key in dict(data.dtypes) if dict(data.dtypes)[key] in ['float64', 'int64']]].pct_change()
returns = data[["AAPL","GOOG","XOM","CVX","VLO","GSPC","FVX"]].pct_change()
```

```python
returns.head()
```

```python
returns = returns[1:]
```

```python
yVars = returns[["GOOG","XOM","CVX","VLO","GSPC","FVX"]]
```

```python
stdYVars = (yVars - yVars .mean()) / (yVars .std())
```

```python
yVars .cov()
yVars .corr()
```

```python
eig_val, eig_vec = np.linalg.eig(stdYVars.cov())
```

```python
eig_pairs = [(np.abs(eig_val[i]), eig_vec[:,i]) for i in range(len(eig_val))]
```

```python
eig_pairs.sort(key=lambda x: x[0], reverse=True)
```

```python
eig_pairs
```

```python
eVector1 = eig_vec[:,0]
eVector2 = eig_vec[:,1]
eVector3 = eig_vec[:,2]
```

```python
pca1 = np.dot(stdYVars,eVector1.T).T
pca1
```

```python
pca2 = np.dot(stdYVars,eVector2.T).T
```

```python
pca3 = np.dot(stdYVars,eVector3.T).T
```

```python
xData = np.vstack([pca1,pca2,pca3]).T
xData
```

```python
stdReturns = (returns["AAPL"] - returns["AAPL"].mean()) / (returns["AAPL"].std())
```

```python
yData = (returns["AAPL"] - returns["AAPL"].mean()) / (returns["AAPL"].std())
```
